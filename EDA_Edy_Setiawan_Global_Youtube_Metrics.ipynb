```python
# mencari apakah ada tahun yang janggal, maka dilakukan teknik Interquartile Range (IQR) yang digunakan untuk mengidentifikasi outlier pada tahun

df = dataset.copy() #copy dataset to new dataframe 'df'

#mencari Q1 dan Q3 serta IQR
Q1 = df['created_year'].quantile(0.25)
Q3 = df['created_year'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = df[(df['created_year'] < lower_bound) | (df['created_year'] > upper_bound)]
print("Outliers:")
print(outliers['created_year'])
```

```python
# Berapa standar deviasi dari jumlah subscribers di setiap negara untuk mengukur variasinya?

std_dev_by_country = dfclean.groupby('country').agg({'subscribers': 'std'}).sort_values(by='subscribers', ascending=False).round(0)
print('\n**Standar Deviasi dari Top 20 negara dengan subscribers tertinggi**\n')
print(std_dev_by_country.head(20)) #ambil top 20 negara dengan value std tertinggi
```

```python
# Ambil standar deviasi untuk top 20 negara dengan subscribers tertinggi berdasarkan index "top_20_countries"
std_dev_top_20 = std_dev_by_country.loc[top_20_countries.index]

# Visualiasi
plt.figure(figsize=(15, 10))
sns.barplot(x=top_20_countries['subscribers'], y=top_20_countries.index)
plt.title('Top 20 Negara dengan Jumlah Subscribers Tertinggi')
plt.xlabel('Jumlah Subscribers (in Millions)')
plt.ylabel('Negara')
plt.show()

plt.figure(figsize=(15, 10))
sns.barplot(x=std_dev_top_20['subscribers'], y=std_dev_top_20.index)
plt.title('Standar Deviasi Jumlah Subscribers untuk Top 20 Negara')
plt.xlabel('Standar Deviasi Subscribers (in Millions)')
plt.ylabel('Negara')
plt.show()
```

Ini menunjukkan bahwa pasar Amerika dan India sangat penting bagi konten kreator YouTube, dan mungkin merefleksikan sejumlah kombinasi dari jumlah populasi, penetrasi internet, dan konsumsi media digital di kedua negara tersebut.

#### Q2. Kategori Populer Berdasarkan Negara:

Q: Kategori YouTube apa yang paling populer di tiap negara berdasarkan jumlah subscribers? Dan adakah kategori tertentu yang mendominasi di sejumlah besar negara?

P: Mengidentifikasi kategori YouTube yang paling populer di setiap negara untuk memberikan wawasan tentang preferensi lokal dan melihat tren regional atau global yang mungkin ada.



```python
# Grouping Data: Kita harus mengelompokkan data berdasarkan negara ('country') dan kategori ('category'). Selanjutnya, kita hitung jumlah total subscribers untuk setiap kombinasi negara dan kategori.
# Mengelompokkan berdasarkan 'country' dan 'category' dan menghitung jumlah total subscribers
grouped_kategori = dfclean.groupby(['country', 'category'])['subscribers'].sum().reset_index().sort_values(by='subscribers', ascending=False)

# Menemukan kategori dengan jumlah subscribers terbanyak untuk setiap negara
idx = grouped_kategori.groupby('country')['subscribers'].transform(max) == grouped_kategori['subscribers']
most_popular_categories = grouped_kategori[idx]

print(most_popular_categories)
```

#### Q3.	Durasi Eksistensi vs Kesuksesan Channel: 

Q:  Apakah channel YouTube yang sudah lama dibuat cenderung lebih sukses daripada yang baru, jika kita mempertimbangkan jumlah subscribers, views, dan engagement rate?

P:  Untuk mendapatkan gambaran yang lebih komprehensif tentang hubungan antara usia channel dan kesuksesannya.


```python
# Mencari Engagement rate, dihitung menggunakan formula sederhana "(video_views / subscribers) * 100" untuk mendapatkan persentase.

dfclean['engagement_rate'] = (dfclean['video views'] / dfclean['subscribers']) * 100

grouped_rate = dfclean.groupby('created_year').agg({
    'subscribers': 'mean',
    'video views': 'mean',
    'engagement_rate': 'mean'
}).reset_index()

grouped_rate.head(20)
```

```python
# Plot untuk rata-rata subscribers
plt.figure(figsize=(12, 12))
plt.subplot(3, 1, 1)
plt.plot(grouped_rate['created_year'], grouped_rate['subscribers'], marker='o', color='blue', label='Subscribers')
plt.title('Rata-Rata Subscribers Berdasarkan Tahun Pembuatan Channel')
plt.ylabel('Avg. Subscribers (in Millions)')
plt.grid(True)
plt.legend()

# Plot untuk rata-rata video views
plt.subplot(3, 1, 2)
plt.plot(grouped_rate['created_year'], grouped_rate['video views'], marker='o', color='green', label='Video Views')
plt.title('Rata-Rata Video Views Berdasarkan Tahun Pembuatan Channel')
plt.ylabel('Avg. Video Views (in Millions)')
plt.grid(True)
plt.legend()

# Plot untuk rata-rata engagement rate
plt.subplot(3, 1, 3)
plt.plot(grouped_rate['created_year'], grouped_rate['engagement_rate'], marker='o', color='red', label='Engagement Rate (%)')
plt.title('Rata-Rata Engagement Rate Berdasarkan Tahun Pembuatan Channel')
plt.xlabel('Avg. Tahun Pembuatan')
plt.ylabel('Engagement Rate (%)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
dfclean['current_date'] = pd.Timestamp.now()  # Kolom dengan tanggal saat ini
dfclean['creation_date'] = pd.to_datetime(dfclean['creation_date'])  # Pastikan 'creation_date' dalam format datetime
dfclean['channel_age'] = (dfclean['current_date'] - dfclean['creation_date']).dt.days / 365  # Menghitung usia dalam tahun

# metode DataFrame yang menghitung korelasi (dalam hal ini, korelasi Pearson) antara setiap pasangan kolom dalam dataframe yang dipilih
# Korelasi berkisar antara -1 sampai 1
correlation_matrix = dfclean[["channel_age", "subscribers", "video views", "engagement_rate"]].corr()

correlation_matrix
```

```python
# vsualiasi heatmap
plt.figure(figsize=(10,7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Korelasi antara Usia Channel dengan Indikator Kesuksesan")
plt.show()
```

**Analisis:**

- Meskipun usia channel memiliki korelasi positif dengan jumlah subscribers dan video views, hubungannya cukup lemah. Ini berarti ada faktor lain selain usia channel yang mempengaruhi kesuksesan suatu channel YouTube.

- Jumlah subscribers dan video views memiliki hubungan yang kuat, yang menunjukkan bahwa mereka adalah indikator kesuksesan yang saling berkaitan.

- Engagement rate, sementara itu, lebih berkorelasi dengan jumlah views daripada jumlah subscribers.

#### Q5.	Pengaruh Urbanisasi terhadap Kesuksesan Channel:

Q : Apakah ada hubungan antara tingkat urbanisasi suatu negara dengan kesuksesan saluran YouTube-nya, jika kesuksesan diukur berdasarkan kombinasi dari jumlah subscribers dan video views?

P : Visualisasi distribusi geografis dari saluran YouTube berdasarkan jumlah subscribers.


**Hipotesis Nol (H0)**: Tidak ada perbedaan signifikan dalam kesuksesan saluran YouTube antara negara dengan tingkat urbanisasi tinggi dan rendah.

**Hipotesis Alternatif (H1)**: Ada perbedaan signifikan dalam kesuksesan saluran YouTube antara negara dengan tingkat urbanisasi tinggi dan rendah.


```python
# Normalisasi kolom subscribers dan video views agar memiliki rentang yang sama (0 sampai 1)

dfclean['norm_subscribers'] = (dfclean['subscribers'] - dfclean['subscribers'].min()) / (dfclean['subscribers'].max() - dfclean['subscribers'].min())
dfclean['norm_video_views'] = (dfclean['video views'] - dfclean['video views'].min()) / (dfclean['video views'].max() - dfclean['video views'].min())

# Buat skor kesuksesan berdasarkan rata-rata normalisasi dari subscribers dan video views
dfclean['success_score'] = (dfclean['norm_subscribers'] + dfclean['norm_video_views']) / 2

# Tentukan saluran sebagai "sukses" jika skor kesuksesannya berada di atas Q3
q3_score = dfclean['success_score'].quantile(0.75)
dfclean['status_success_score'] = ['sukses' if x > q3_score else 'biasa' for x in dfclean['success_score']]

# Tampilkan jumlah saluran yang dianggap "sukses"
sukses_score_rows = dfclean[dfclean['status_success_score'] == 'sukses']
print(sukses_score_rows.shape[0])
```

```python
median_urbanization = dfclean['urban_population'].median()
high_urban = dfclean[dfclean['urban_population'] > median_urbanization]
low_urban = dfclean[dfclean['urban_population'] <= median_urbanization]

# uji t (t-test) untuk membandingkan rata-rata skor kesuksesan dari kedua kelompok.
from scipy.stats import ttest_ind

t_stat, p_val = ttest_ind(high_urban['success_score'], low_urban['success_score'])

print("T-stat:", t_stat)
print("P-value:", p_val)
```

```python
# Mengatur style
sns.set_style("whitegrid")

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=dfclean['urban_population'], y=dfclean['success_score'])
plt.title('Scatter plot of Success Score vs. Urban Population')
plt.xlabel('Urban Population (in Millions)')
plt.ylabel('Success Score')
plt.show()
```

**Analisi:**

•	Dari hasil analisis, menemukan bahwa Channel YouTube dari negara dengan tingkat urbanisasi yang lebih tinggi cenderung memiliki skor kesuksesan yang lebih baik. Namun, perbedaan ini tidak statistik signifikan pada level 0.05, dengan p-value sebesar 0.0501, yang hampir mencapai ambang batas signifikansi, sehingga H0 tidak dapat ditolak.

#### Q6.	: Hubungan antara pertumbuhan subscribers dan video views pada channel YouTube

Q : Apakah ada korelasi positif antara pertumbuhan jumlah subscribers dan pertumbuhan video views dalam 30 hari terakhir?

P : Mengkaji apakah peningkatan jumlah pelanggan dalam 30 hari terakhir berpengaruh terhadap peningkatan jumlah video views pada periode yang sama.


**Hipotesis Nol (H0)**: Tidak ada korelasi antara pertumbuhan jumlah subscribers dan pertumbuhan video views dalam 30 hari terakhir.

**Hipotesis Alternatif (H1)**: Ada korelasi antara pertumbuhan jumlah subscribers dan pertumbuhan video views dalam 30 hari terakhir.


```python
# Hitung Korelasi Pearson dan Tes Signifikansi
correlation, p_value = pearsonr(dfclean['subscribers_for_last_30_days'], dfclean['video_views_for_the_last_30_days'])
print(f"Korelasi Pearson: {correlation:.2f}")

if p_value < 0.05:
    print(f"P-value: {p_value:.5f}. Hasilnya signifikan, sehingga H0 ditolak.")
else:
    print(f"P-value: {p_value:.8f}. Hasilnya tidak signifikan, sehingga H0 tidak dapat ditolak.")

# Visualisasikan dengan Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dfclean, x='subscribers_for_last_30_days', y='video_views_for_the_last_30_days')
plt.title('Scatter Plot of Growth in Subscribers vs. Growth in Video Views (Last 30 days)')
plt.xlabel('Growth in Subscribers')
plt.ylabel('Growth in Video Views')
plt.show()
```

**Analisi:**

•	Terdapat korelasi positif yang signifikan antara pertumbuhan jumlah subscribers dan pertumbuhan video views dalam 30 hari terakhir dengan koefisien korelasi sebesar 0.62. Ini menunjukkan bahwa saat jumlah subscribers meningkat, pertumbuhan video views juga cenderung meningkat, dan sebaliknya. Oleh karena p-value mendekati 0 (kurang dari batas umum 0.05), maka ada bukti kuat bahwa korelasi ini signifikan dan bukan karena kebetulan. Sehingga, hipotesis nol (H0) yang menyatakan tidak ada korelasi ditolak.

#### Q7. : Pengaruh Tingkat Pengangguran Negara Asal Terhadap Kesuksesan Channel YouTube. 

Q : Apakah ada korelasi antara tingkat pengangguran negara asal dengan kesuksesan channel YouTube (dinyatakan dalam jumlah subscribers)?

P : Mengetahui apakah individu dari negara dengan tingkat pengangguran yang tinggi lebih cenderung beralih ke YouTube sebagai sumber pendapatan, yang tercermin dari jumlah subscribers.


**Hipotesis Nol (H0)**: Tidak ada korelasi antara tingkat pengangguran negara asal dan jumlah subscribers channel YouTube.

**Hipotesis Alternatif (H1)**: Ada korelasi antara tingkat pengangguran negara asal dan jumlah subscribers channel YouTube.



```python
# Menghitung Korelasi Pearson dan Tes Signifikansi
correlation, p_value = pearsonr(dfclean['unemployment rate'], dfclean['subscribers'])
print(f"Korelasi Pearson: {correlation:.2f}")

if p_value < 0.05:
    print(f"P-value: {p_value:.5f}. Hasilnya signifikan, sehingga H0 ditolak.")
else:
    print(f"P-value: {p_value:.5f}. Hasilnya tidak signifikan, sehingga H0 tidak dapat ditolak.")

# Visualisasikan dengan Regression Plot
# Plot regresi dapat membantu visualisasi baik dispersi titik data maupun garis regresi linear, yang membantu memahami hubungan linier antar variabel.
sns.regplot(data=dfclean, x='unemployment rate', y='subscribers')
plt.xlabel('Tingkat Pengangguran')
plt.ylabel('Jumlah Subscriber (in Millions)')
plt.title('Plot Regresi Tingkat Pengangguran vs. Jumlah Subscriber')
plt.show()
```

P-value sebesar 0.94787 menunjukkan bahwa korelasi yang ditemukan tidak signifikan secara statistik. Oleh karena itu, hipotesis nol (H0) yang menyatakan bahwa tidak ada korelasi antara tingkat pengangguran negara asal dan jumlah subscribers channel YouTube tidak dapat ditolak.
